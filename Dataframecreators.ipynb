```python
#import packages
import pandas as pd
import os
import datetime 
import seaborn as sns
from matplotlib import dates as mpl_dates
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
#data preparation after webscraping data 
df = pd.read_csv("Combined.csv", index_col = 0)
df['Swim Time DT'] = pd.to_datetime(df['Swim Time'])
df['Bike Time DT'] = pd.to_datetime(df['Bike Time'])
df['Run Time DT'] = pd.to_datetime(df['Run Time'])
df['Total Time DT'] = pd.to_datetime(df['Total Time'])

df_am = df[(df['Sex'] != 'X') & (df['Sex'] != 'U' ) & (df['Age Group'] != 'PRO' ) & (df['Age Group'] != 'PC' )] #ameatur atheletes
df_am.head()
```

```python
#convert to necessary datetime objects 

df_am['Swim Time Obj'] = df_am['Swim Time']/np.timedelta64(1, 'h')
df_am['Bike Time Obj'] = df_am['Bike Time']/np.timedelta64(1, 'h')
df_am['Run Time Obj'] = df_am['Run Time']/np.timedelta64(1, 'h')
df_am['Total Time Obj'] = df_am['Total Time']/np.timedelta64(1, 'h')

#df_am['Swim Time Obj'] = np.median(df_am['Swim Time'].dt.time)

quant_val_swim = df_am['Swim Time Obj'].quantile(q=[0.25, .75], interpolation = 'higher')

quant_val_bike = df_am['Bike Time Obj'].quantile(q=[0.25, .75], interpolation = 'higher')

quant_val_run = df_am['Run Time Obj'].quantile(q=[0.25, .75], interpolation = 'higher')


x25_swim = df_am.loc[df_am['Swim Time Obj'] == quant_val_swim[.25]].head(1) ['Swim Time DT']

x75_swim = df_am.loc[df_am['Swim Time Obj'] == quant_val_swim[.75]].head(1) ['Swim Time DT']

x25_bike = df_am.loc[df_am['Bike Time Obj'] == quant_val_bike[.25]].head(1) ['Bike Time DT']

x75_bike = df_am.loc[df_am['Bike Time Obj'] == quant_val_bike[.75]].head(1) ['Bike Time DT']

x25_run = df_am.loc[df_am['Run Time Obj'] == quant_val_run[.25]].head(1) ['Run Time DT']

x75_run = df_am.loc[df_am['Run Time Obj'] == quant_val_run[.75]].head(1) ['Run Time DT']


xmedian_s = np.median(df_am['Swim Time Obj'])

xmedian_swim = df_am.loc[df_am['Swim Time Obj'] == xmedian_s].head(1) ['Swim Time DT']

xmedian_b = np.median(df_am['Bike Time Obj'])

xmedian_bike = df_am.loc[df_am['Bike Time Obj'] == xmedian_b].head(1) ['Bike Time DT']

xmedian_r = np.median(df_am['Run Time Obj'])

xmedian_run = df_am.loc[df_am['Run Time Obj'] == xmedian_r].head(1) ['Run Time DT']
```
